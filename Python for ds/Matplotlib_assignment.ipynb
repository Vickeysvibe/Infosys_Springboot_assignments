# Problem Statement

Given the following values for x and y, use Matplotlib to plot the figure shown below:

1. import numpy as np

2. x=np.arange(0.0, 5.0, 0.1)

3. y=np.cos(2* np.pi*x)*np.exp(-x)



# Solution

```python
import numpy as np
import matplotlib.pyplot as plt
x= np.arange(0.0, 5.0, 0.1)
y= np.cos(2*np.pi*x)*np.exp(-x)
```

```python
fig, ax = plt.subplots()
ax.plot(x, y, marker='o', linestyle='-', color='black',markersize=10, alpha=0.5, markerfacecolor='cyan')
plt.title('Main heading')
plt.grid(True)
plt.annotate('θ = 60°', xy=(1, 0.5), xytext=(1.5, 0.7))
plt.annotate('2nd crest', xy=(2, 0.2), xytext=(3, 0.4), arrowprops=dict(facecolor='green', shrink=0.05),fontsize=12)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the data
x = np.arange(0.0, 5.0, 0.1)
y = np.cos(2 * np.pi * x) * np.exp(-x)

# Create the plot
plt.figure()
plt.plot(x, y)

# Customize the plot
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Damped Cosine Wave')
plt.grid(True)

# Display the plot
plt.show()
```
